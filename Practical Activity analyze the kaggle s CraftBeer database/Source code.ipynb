### 1) Importando pandas

```python
import pandas as pd
```

### 2) Leitura dos dados

```python
data_beers = pd.read_csv('files/beers.csv')
data_brew = pd.read_csv('files/breweries.csv')
```

### 3) União das duas bases de dados

```python
##### Padronizar a coluna brewery_id em breweries.csv
data_brew.columns = ['brewery_id', 'name', 'city', 'state']
data_beers.columns = ['Unnamed: 0', 'abv', 'ibu', 'id', 'name_beer', 'style', 'brewery_id', 'ounces']
```

```python
data_brew
```

```python
##### Cria um único DataFrame com as duas bases
df = pd.merge(data_beers, data_brew, on=['brewery_id'])
```

```python
df
```

# Responder as seguintes questões:

### i) Qual o estado que produz mais cervejas ?

```python
state = df.groupby('state').name_beer.count().sort_values(ascending=False)
state.plot()
# OR
#state
```

### ii) Qual o tipo (style) mais produzida nos EUA ?

```python
style = df.groupby('style').style.count().sort_values(ascending=False)
style.head(3).plot()
# OR
#style
```

### iii) Qual a fabricante (Brewery) que tem maior diversidade de tipos de cervejas?

```python
pd.set_option('display.max_rows', None)
# Eliminando valores nullos
aux = df[df['style'].isnull() == False]
# Gerando um novo data frame 
aux2 = pd.DataFrame()

# Preenchendo o data frame somente com as colunas de interesse, duas delas formatadas para não ter problema com relação ao nome
aux2['original_name'] = aux['name']
aux2['original_style'] = aux['style']
aux2['name'] = aux['name'].str.replace(' ', '').str.upper()
aux2['style'] = aux['style'].str.replace(' ', '').str.upper()

# Calculando a quantidade de vezes que cada tipo de cerveja aparece para cada fabricante
aux3 = aux2.groupby('name')['style'].value_counts()
# Agrupando por fabricante e calculando a quantidade de registro que esse possui considerando o resultado do procedimento anterior
aux4 = aux3.groupby('name').count().sort_values(ascending=False)

dicionario = dict(aux4)

#dicionario

for i in dicionario:
    print(aux2[aux2['name'] == i]['original_name'].unique())
    break

# Para testar se os dados estão corretos (basta informar o nome formatado do frabricante)
#len(aux2[aux2['name'] == 'LEWISANDCLARKBREWINGCOMPANY']['style'].unique())
```
